```python
# Banking system

class BankAccount:
    #Constructor
    def __init__(self,account_holder,initial_balance):
        self.account_holder = account_holder
        self.balance = initial_balance
    #Amount withdrawal
    def withdraw(self,amount):
        self.balance -= amount
    #Amount deposit
    def deposit(self,amount):
        self.balance += amount
    #Amount transfer
    def transfer(self,other_account,amount):
        self.balance -= amount
        other_account.deposit(amount)
    def __repr__(self):
        return f"Account of {self.account_holder} has amount {self.balance}"



acc1 = BankAccount("Account1",2500)
acc2 = BankAccount("Account2",2000)

print("Before withdrawals or deposits, initial details are : ")
print(acc1)
print(acc2)

acc1.withdraw(500)
acc2.deposit(200)

print("\nAfter withdrawal of 500 from acc1 and deposit of 200 from acc2 : ")
print(acc1)
print(acc2)

print("\nAfter transfering 1000 from acc1 to acc2 : ")
acc1.transfer(acc2,1000)
print(acc1)
print(acc2)
```

```python
# Implementing fraction

class Fraction:
    def __init__(self,numerator,denominator):
        self.numerator = numerator
        self.denominator = denominator

    def _find_gcd(self,a,b):
        if a == 0:
            return b
        if b == 0:
            return a
        if a < b:
            a,b = b,a
        return self._find_gcd(b,a%b)

    def _simplify_fraction(self):
        gcd = 0
        if self.numerator >= 0:
            gcd = self._find_gcd(self.numerator,self.denominator)
        else:
            gcd = self._find_gcd(-self.numerator,self.denominator)
        return Fraction(self.numerator//gcd,self.denominator//gcd)

    def add(self,other):
        new_numerator = self.numerator*other.denominator + self.denominator*other.numerator
        new_denominator = self.denominator*other.denominator
        self.numerator = new_numerator
        self.denominator = new_denominator
        return self._simplify_fraction()

    def subtract(self,other):
        new_numerator = self.numerator*other.denominator-self.denominator*other.numerator
        new_denominator = self.denominator*other.denominator
        self.numerator = new_numerator
        self.denominator = new_denominator
        return self._simplify_fraction()

    def multiply(self,other):
        new_numerator = self.numerator*other.numerator
        new_denominator = self.denominator*other.denominator
        self.numerator = new_numerator
        self.denominator = new_denominator 
        return self._simplify_fraction()

    def __repr__(self):
        return f"{self.numerator}/{self.denominator}"


fraction1 = Fraction(2,4)
fraction2 = Fraction(3,4)
print(f"Sum of 2/4 and 3/4 is : {fraction1.add(fraction2)}")

# Since fraction1 value changed previously
fraction1 = Fraction(2,4)
print(f"Difference between 2/4 and 3/4 is : {fraction1.subtract(fraction2)}")

fraction1 = Fraction(2,4)
print(f"Product of 2/4 and 3/4 is : {fraction1.multiply(fraction2)}")
```
